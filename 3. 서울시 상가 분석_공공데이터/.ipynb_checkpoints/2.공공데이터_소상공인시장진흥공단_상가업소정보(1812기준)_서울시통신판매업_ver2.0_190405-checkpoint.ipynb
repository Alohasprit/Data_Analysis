```python
col1=(['상호명','상권업종대분류명','상권업종중분류명','상권업종소분류명','표준산업분류명','시도명',
     '시군구명','행정동명','법정동명','위도','경도'])
```

```python
data1=data1[col1]
data1
```

```python
#서울특별시에 위치한 상권업종중소불류명 = 핸드폰 소매점 
s_data=data1.loc[(data1['상권업종소분류명']=='핸드폰소매') & (data1['시도명']=='서울특별시')]
```

```python
s_data.head(3)
```

```python
cols=['상호명','시군구명','법정동명','위도','경도']
ss_data=s_data[cols]
ss_data.shape
```

```python
ss_data.head(3)
```

```python
ss_data.iloc[3,0]
```

```python
#위도, 경도 기준 지도를 세팅하고, 
map_basic = fl.Map(
    location=[37.55, 126.971331],
    zoom_start=11
)

#지도위에 마커 생성 및 마커 정보 생성

for i in range(0,500):
    fl.Marker(
        [ss_data.iloc[i,3], ss_data.iloc[i,4]],
        popup=ss_data.iloc[i,2],
        tooltip =ss_data.iloc[i,0],
        icon=fl.Icon(color='red', icon='info-sign')
    ).add_to(map_basic)

map_basic
```
