### 위도 경도에 찍어보자
[Folium 라이브러리 설명](https://python-visualization.github.io/folium/)

```python
#위도, 경도 기준 지도를 세팅하고, 
map_basic = fl.Map(
    location=[35.149792, 129.004052],
    tiles='Stamen Terrain',
    zoom_start=11
)

#지도위에 마커 생성 및 마커 정보 생성
fl.Marker(
    [35.149792, 129.004052],
    popup='<i>상세주소</i>',
    tooltip ='상호명:하자하자다이어트',
    icon=fl.Icon(color='red', icon='info-sign')
).add_to(map_basic)

map_basic
```
